# NUMPY

````python
import tracemalloc
import numpy as np
import matplotlib.pyplot as plt
import psutil
import time

def Numpy():
  list=[[1,23,45,67],[2,3,43,5]]
  print("Transpose of list:\n",np.array(list).transpose())
  print("Array1:\n",np.arange(0,101,20))
  print('Array2:\n',np.linspace(0,1,21))
  np.random.seed(101)
  print("Array3:\n",np.random.rand(2,3))

tracemalloc.start()
start = time.time()
Numpy()
tupl=(tracemalloc.get_traced_memory()) #he output is given in form of (current, peak),i.e, current memory is the memory the code is currently using and peak memory is the maximum space the program used while executing.
tracemalloc.stop()
end = time.time()

data={
    'Current Memory Using':tupl[0],
    'Peak Memory Used':tupl[1]}
data1={
    'RAM memory % used':psutil.virtual_memory()[2],
    'RAM Used (GB)':psutil.virtual_memory()[3]/1000000000
}
time={
    'Time_elapsed':end-start
}

time_library['numpy_time_elapsed']=time['Time_elapsed']
memory_library['numpy_current_memory']=data['Current Memory Using']
RAM_library['numpy_RAM_USED(GB)']=data1['RAM Used (GB)']




print("\n----------------------------------------------------------------------------------------------\n")
print("Traced Memory used by function Numpy:",data)

items = list(data.keys())
values = list(data.values())

items1=list(data1.keys())
values1=list(data1.values())

items2=list(time.keys())
values2=list(time.values())

plt.subplot(1, 3, 1)
plt.bar(items, values, color='red');
plt.xticks(rotation=90);
plt.title('Traced Memory');

print("Virtual Memory used by function Numpy:",data1)
plt.subplot(1, 3, 2)
plt.bar(items1, values1, color='green');
plt.xticks(rotation=90);
plt.title('Virtual Memory');

print("Time",time)
plt.subplot(1,3,3)
plt.bar(items2,values2,color='skyblue');
plt.xticks(rotation=90)
plt.title('time');
````

# SCIPY
